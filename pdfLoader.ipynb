Create loader and load pdf

```python
# load document
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("Mahmoud Abuelhasaan Full Stack.pdf")
documents = loader.load()
```

```python
print(f"Number of pages: {len(documents)}")
```

#split fils to chanks 

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
)

chunks = text_splitter.split_documents(documents)
```

```python
len(chunks)
```

```python
chunks[0].page_content
```

```python
chunks[1].page_content
```

save chunks into vector db

```python
from langchain_ollama import OllamaEmbeddings
```

```python
em_llm = OllamaEmbeddings(model="llama3:8b")
```

```python
sample_text = 'my name is mahmoud '
em = em_llm.embed_query(sample_text)
em
```

```python
len(em)
```

###chroma

```python
from langchain_chroma import Chroma
```

```python
vector_db = Chroma.from_documents(
    documents=chunks,
    embedding=em_llm,
    persist_directory="vector_db/cv_db"
)
```

```python
# create retriever
retriever = vector_db.as_retriever(search_kwargs={"k": 3})
retriever
```

```python
# test retriever
q="What is Work Experience section in my CV?"
ret_docs = retriever.invoke(q)
```

```python
for i , doc in enumerate(ret_docs):
    print(f"Document {i}:")
    print(doc.page_content)
    print("Metadata:", doc.metadata)
    print("-" * 40)
```

```python
from langchain_ollama.llms import OllamaLLM
from langchain.chains import RetrievalQA

# Instantiate the Ollama model
llm = OllamaLLM(model="llama3:8b")

# Create RAG chain using RetrievalQA
rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,  # Make sure retriever is defined
    return_source_documents=True
)

# Run the RAG chain
result = rag_chain.invoke({"query": "What is my name?"})
print(result)
```
